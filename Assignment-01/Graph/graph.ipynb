```python
from matplotlib import pyplot as plt

x = [12, 4, 3, 3, 1]
plt.plot(x)
plt.show()
```

```python
x = [4, 1, 9, 12, 21]
y = [12, 10, 21, 21, 17] 

plt.bar(x, y, width=0.5)
plt.show()
```

```python
plt.xlabel(" X values")
plt.ylabel(" Scores")
plt.title(" Sample Bar Graph")
label = ['Math', 'English', 'Comp', 'Physics', 'Alg']
plt.bar(x, y, tick_label = label, color = ['green', 'blue'])
```
